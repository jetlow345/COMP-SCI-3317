```python
# load libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data
data = pd.read_csv("workshop2.csv")
```

```python
# See data
data
```

```python
# Check first rows
data.tail()
```

```python
# Check description of data
data.describe()
```

```python
# Check data info
data.info()
```

```python
# Remove a column 
data = data.drop('ocean_proximity',axis = 1)
```

```python
# Convert column 'total_rooms' to numeric
data['total_rooms'] =  pd.to_numeric(data['total_rooms'],errors = 'coerce')
```

```python
# Remove rows with nan values (missing values)
data = data.dropna()
```

```python
# Plot one variable
plt.plot(data['total_rooms'],'o')
```

```python
# Plot multiple variables
for col in data.columns:
    plt.plot(data[col],'o')
    plt.title(col)
    plt.show()
```

```python
# Checking data, we identify a 'potential' wrong value of latitude, which value is over 350. We can check this sample.
data.loc[data['latitude'] > 350,:]
```

```python
# We can replace this value for nan
data.loc[data['latitude'] > 350,'latitude'] = np.nan
```

```python
# We can plot the vaiable latitude again and corroborate that value over 350 was removed
plt.plot(data['latitude'],'o')
```

```python
# As we introduced a new nan value, we need to remove the entire row with the missing value
data = data.dropna()
```

```python
# Using data describe we can check that our variables have the same number of correct values (checking the row 'count')
data.describe()
```

```python
# Now we can split our data in y and X
y = data['median_house_value']
X = data.loc[:,'longitude':'median_income']
```

```python
# Now, we can split our dataset in train and test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)
```

```python
# Now  we can implement a linear regression model 
from sklearn.linear_model import LinearRegression

# Create a model object
model = LinearRegression()
# Train a model
model.fit(X_train,y_train)

# We can do prediction in train set
model.predict(X_train)
```

```python
# We can do prediction in test set
model.predict(X_test)
```
